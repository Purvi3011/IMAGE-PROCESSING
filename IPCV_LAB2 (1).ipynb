```python
#THRESHOLDING
img2 = cv2.imread("img 1.jpg",0)

```

```python
x,y = img2.shape
```

```python
print(x,y)
```

```python
import numpy as np
z = np.zeros((x,y))
```

```python
for i in range(0,x):
    for j in range(0,y):
        if(img2[i][j]>150):
            z[i][j]=255;
        else:
            z[i][j]=0;
output = np.hstack((img2,z))
plt.imshow(output,cmap="gray")
```

```python
for i in range(0,x):
    for j in range(0,y):
        if(img2[i][j]>200):
            z[i][j]=255;
        else:
            z[i][j]=0;
output = np.hstack((img2,z))
plt.imshow(output,cmap="gray")
```

```python
#GREY LEVEL SLICING WITHOUT BACKGROUND
img3 = cv2.imread("images.jpg",0)
```

```python
p,q = img3.shape
```

```python
print(p,q)
```

```python
z = np.zeros((p,q))
```

```python
for p in range(0,p):
    for q in range(0,q):
        if(img3[p][q]>=150 and img3[p][q] < 200):
            z[p][q]=255;
        else:
            z[p][q]=0;
output = np.hstack((img3,z))
plt.imshow(output,cmap="gray")
```

```python
#GREY LEVEL SLICING WITH BACKGROUND

for p in range(0,p):
    for q in range(0,q):
        if(img3[p][q]>=150 and img3[p][q] < 200):
            z[p][q]=255;
        else:
            z[p][q]=img3[p][q];
output = np.hstack((img3,z))
plt.imshow(output,cmap="gray")
```
