```python
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import numpy as np
import math
import random
```

```python
img = cv2.imread('/content/img1.jfif')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img , cmap = "gray")
```

```python
#printing shapes
x , y = (img.shape)
print(x)
print(y)
```

```python
#high pass filter
filter = np.ones([3, 3], dtype = int)
filter = - filter
filter[1,1] = - 8 * filter[1,1]
print(filter)
```

```python
highPassed_filtered = np.zeros([x, y])

for i in range(1, x-1):
	for j in range(1, y-1):
		temp = (img[i-1, j-1]*filter[0, 0] + img[i-1, j]*filter[0, 1] + 
          img[i-1, j + 1]*filter[0, 2] + img[i, j-1]*filter[1, 0] + 
          img[i, j]*filter[1, 1] + img[i, j + 1]*filter[1, 2] + 
          img[i + 1, j-1]*filter[2, 0] + img[i + 1, j]*filter[2, 1] +
          img[i + 1, j + 1]*filter[2, 2])
		
		highPassed_filtered[i, j]= temp
```

```python
fig=plt.figure(dpi = 150)
fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,2,2)
plt.imshow(highPassed_filtered,cmap='gray')
plt.axis("off")
plt.title("High Pass Filtered")
```

```python
#applying high boost filter
a = 1.8
```

```python
filter = np.ones([3, 3], dtype = int)
filter = - filter
filter[1,1] = -( 8 + a )* filter[1,1]
```

```python
highBoost_filtered = np.zeros([x, y])

for i in range(1, x-1):
	for j in range(1, y-1):
		temp = (img[i-1, j-1]*filter[0, 0] + img[i-1, j]*filter[0, 1] + 
          img[i-1, j + 1]*filter[0, 2] + img[i, j-1]*filter[1, 0] + 
          img[i, j]*filter[1, 1] + img[i, j + 1]*filter[1, 2] + 
          img[i + 1, j-1]*filter[2, 0] + img[i + 1, j]*filter[2, 1] +
          img[i + 1, j + 1]*filter[2, 2])
		
		highBoost_filtered[i, j]= temp
```

```python
fig=plt.figure(dpi = 200)
fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,2,2)
plt.imshow(highBoost_filtered,cmap='gray')
plt.axis("off")
plt.title("High Boost Filtered")
```

**CONCLUSION : Hence I conclude that by applying high pass filter the image appear sharper and by applying High boost filter it emphasize high frequency components representing the image details without eliminating low frequency components**
