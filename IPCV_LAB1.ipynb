```python
import numpy as np 
img2 = cv2.imread('img2.jpg')
img3 = cv2.imread('img3.jpg')
weightedSum = cv2.addWeighted(img2, 0.5, img3, 0.4, 0)
cv2.imshow('Weighted Image', weightedSum)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img3)
```

```python
plt.imshow(weightedSum)
```

```python
sub = cv2.subtract(img2, img3)
```

```python
cv2.imshow('Subtracted Image', sub)
```

```python
plt.imshow(sub)
```

```python
dest_and = cv2.bitwise_and(img2, img3, mask = None)
cv2.imshow('Bitwise And', dest_and)
```

```python
plt.imshow(dest_and)
```

```python
dest_or = cv2.bitwise_or(img2, img3, mask = None)
cv2.imshow('Bitwise OR', dest_or)
```

```python
plt.imshow(dest_or)
```

```python
dest_xor = cv2.bitwise_xor(img2, img3, mask = None)

plt.imshow(dest_xor)
```

```python
dest_not1 = cv2.bitwise_not(img2, mask = None)
dest_not2 = cv2.bitwise_not(img3, mask = None)
```

```python
cv2.imshow('Bitwise NOT on image 1', dest_not1)
cv2.imshow('Bitwise NOT on image 2', dest_not2)
```

```python
plt.imshow(dest_not1)
```

```python
plt.imshow(dest_not2)
```
